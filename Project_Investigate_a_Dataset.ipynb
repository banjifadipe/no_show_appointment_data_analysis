##  What are the determining factors for patients not showing for their appointments?

### Plotting the histogram of the features. Note that the gender, neighbourhood and no_show columns are not plotted because they are not numerical

```python
df.hist(figsize=(10, 8));
```

### The distributions above are skewed to the right

```python
#Masking
noshow= df.query('No_show == "Yes"')
show = df.query('No_show == "No"')
```

```python
show.head()
```

### Comparing the mean age of the patients who didn't appear to those who appeared
Notice that those who appeared happens to be older than those who did not!

```python
df.query('No_show == "Yes"')['Age'].mean(), df.query('No_show == "No"')['Age'].mean()
```

## Plotting the charts of each of the features to determine which affects the appearance of patients at their appointments

```python
def PlotFunction(dfr, colname, color):
    
    counts = dfr.groupby(['No_show']).mean()[colname]
    counts.plot(kind ='bar',title='Appearance by '+colname, color= color, alpha=0.7)
    plt.xlabel(colname, fontsize=18)
    plt.ylabel('Count', fontsize=18);
```

#### Note that No means that patients appeared for their appointments while Yes means that patients did not show up for their appointments!

## Age

```python
PlotFunction(df, 'Age', 'blue')
```

From the above chart, more older patients showed up for their appointments

## Scholarship

```python
PlotFunction(df, 'Scholarship', 'red')
```

From the chart above, being on scholarship does not determine the appearance of patients at their appointments hence, is not a factor for patients showing up as majority of patients on scholarship did not show up.

```python
df.query('No_show == "Yes"')['Handcap'].count(), df.query('No_show == "No"')['Handcap'].count()
```

## Handcap

```python
PlotFunction(df, 'Handcap', 'orange')
```

From the above, more handicap patients made their appointments

## Diabetes

```python
PlotFunction(df, 'Diabetes', '')
```

From, the above, more diabetic patients appeared for their appointments

## Hypertension

```python
PlotFunction(df, 'Hipertension', 'yellow')
```

More hypertensive patients also made their appointments according to the above chart

## Alcoholism

```python
PlotFunction(df, 'Alcoholism', 'black')
```

The chart above shows that same number of alcoholic patients who did not show up showed up, hence Alcoholism is not a determining factor for patients showing up for their appointments

## Received SMS

```python
PlotFunction(df, 'SMS_received', 'green')
```

From the above chart, most patients who received SMS did not show up for their appointments

### Figures below show the distribution of patients that didn't show up and those that showed up for their appointments.

```python
show.groupby(['Age']).mean().plot(kind='hist', figsize=(10,10), title='Patients who showed up')
plt.legend();
```
